```python
import pandas as pd

df = pd.read_csv('/Users/josiahwu29/Desktop/SummerResearchY1/GoalMins.csv')
df['Date'] = pd.to_datetime(df['Date'],dayfirst=True)
print(df)
```

```python
def Cumulative(l):
    newl = []
    for n in range(len(l)):
        newl.append(sum(l[0:n+1]))
    return newl

def gendataframe(t):
    return (df[df['Team']==t]).sort_values(by='Date')

##Calculate minutes from Start of Season
def GoalMinsCumul(t):
    dft = df[df['Team']==t].sort_values(by='Date')
    CumulMins = Cumulative(list(dft['TotalMins']))
    Goals = list(dft['Goals'])
    Final_List = []
    for ind in range(len(Goals)):
        x = str(Goals[ind])
        if x != '-1':
            y = sorted(x.split('*'))
            if ind == 0:
                Final_List.extend([int(z) for z in y])
            else:
                Final_List.extend([int(z)+int(CumulMins[ind-1]) for z in y])
    return sorted(Final_List)
    
## Calculate difference in minutes between each goal
def GoalMinsDiff(t):
    l = GoalMinsCumul(t)
    Newl = [l[0]]
    j = 1
    while j< len(l):
        Newl.append((l[j]-l[j-1]))
        j += 1
    return Newl

Teams = ['Arsenal',
            'Aston Villa',
            'Brentford',
            'Brighton and Hove Albion',
            'Burnley',
            'Chelsea',
            'Crystal Palace',
            'Everton',
            'Leeds United',
            'Leicester City',
            'Liverpool',
            'Manchester City',
            'Manchester United',
            'Newcastle United',
            'Norwich City',
            'Southampton',
            'Tottenham Hotspur',
            'Watford',
            'West Ham United',
            'Wolverhampton Wanderers']

## Calculate Decimal of Goals scored
def GoalMinsFract(t):
    dft = df[df['Team']==t].sort_values(by='Date')
    Goals = list(dft['Goals'])
    TotalMins = list(dft['TotalMins'])
    Final_List = []
    for ind in range(len(Goals)):
        x = str(Goals[ind])
        tot = int(TotalMins[ind])
        if x != '-1':
            y = sorted(x.split('*'))
            for j in range(len(y)):
                Final_List.append(int(y[j])/tot)
    return Final_List

## Calculate Decimal of Goals scored cumulatively
def GoalMinsFractCumul(t):
    dft = df[df['Team']==t].sort_values(by='Date')
    Goals = list(dft['Goals'])
    TotalMins = list(dft['TotalMins'])
    Final_List = []
    for ind in range(len(Goals)):
        x = str(Goals[ind])
        tot = int(TotalMins[ind])
        if x != '-1':
            y = sorted(x.split('*'))
            for j in range(len(y)):
                Final_List.append(ind+int(y[j])/tot)
    return Final_List


print(gendataframe(Teams[0]))
print(GoalMinsCumul(Teams[0]))
print(GoalMinsFract(Teams[0]))
print(GoalMinsFractCumul(Teams[0]))
```

```python
## Generate Dataframes for Cumulative and Difference in Goal Minutes
DiffDF_List = []
CumulDF_List = []
FractDF_List = []
FractCumul_List = []
for n in range(20):
    t = Teams[n]
    diff = GoalMinsDiff(t)
    cumul = GoalMinsCumul(t)
    fract = GoalMinsFract(t)
    fractc = GoalMinsFractCumul(t)
    for i in range(len(diff)):
        DiffDF_List.append([t, int(diff[i])])
    for j in range(len(cumul)):
        CumulDF_List.append([t,int(cumul[j])])
    for k in range(len(fract)):
        FractDF_List.append([t,round(fract[k],4)])
    for q in range(len(fractc)):
        FractCumul_List.append([t,round(fractc[q],4)])

DiffDF = pd.DataFrame(DiffDF_List, columns=['Team','Min'])
CumulDF = pd.DataFrame(CumulDF_List, columns=['Team','Min'])
FractDF = pd.DataFrame(FractDF_List, columns=['Team','Decimal'])
FractCumulDF = pd.DataFrame(FractCumul_List, columns=['Team','Decimal'])

print(DiffDF)
print(CumulDF)
print(FractDF)
print(FractCumulDF)

DiffDF.to_csv('GoalMinsDiff.csv')
CumulDF.to_csv('GoalMinsCumul.csv')
FractDF.to_csv('GoalMinsFract.csv')
FractCumulDF.to_csv('GoalMinsFractCumul.csv')
```
